### Exercici 4

Extracte de la documentació de NumPy:

> In order to broadcast, the size of the trailing axes for both arrays in an operation must either be the same size or one of them must be one.

**Exemple A: Broadcasting de matrius amb eixos de la mateixa forma (3,4)**

```python
arr3 = rng.integers(0, 100, 12).reshape(3, 4)
arr3
```

```python
arr4 = rng.random(12).reshape(3, 4)
arr4
```

```python
arr3 * arr4 
```

**Exemple B: Broadcasting d'una matriu 3D (3, 4, 2) amb un vector.**

```python
arr5 = rng.integers(0, 100, 24).reshape(3, 4, 2)
arr5
```

```python
arr5 % 2
```
